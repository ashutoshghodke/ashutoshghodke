```python
#basic x y and plot it
x=[2019,2020,2021,2022]
y=[10,50,15,145]
plt.title("Turn Over")
plt.xlabel("years")
plt.ylabel("INR")
plt.plot(x,y,color="red",marker="o",markersize=10,linewidth=2)
```

```python
#basic x y and plot it
x=[2019,2020,2021,2022]
y=[10,-50,15,-145]
plt.title("Turn Over")
plt.xlabel("years")
plt.ylabel("INR")
plt.bar(x,y,width=0.1)
```

```python
import matplotlib.pyplot as plt
   
Year = [1900,1920,1930,1940,1950,1960,1970,1980,1990,2000,2010,2020]
Unemployment_Rate = [4.5,9.8,20,8,7.2,6.9,7,6.5,6.2,5.5,6.3,3.3]
New_Colors = ['green','blue','purple','brown','teal',"orange",'green','blue','purple','brown','teal',"orange"]
#bar chart with respective label and title as "industrial revolution"\
plt.title('Industrial Revolution')
plt.xlabel('Year')
plt.ylabel('Unemployment_Rate')
plt.grid()
plt.bar(Year,Unemployment_Rate,width=1,color=New_Colors)
plt.plot(Year,Unemployment_Rate,color="yellow")
```
